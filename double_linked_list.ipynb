```python
# double_linked_list
class Node:
    def __init__(self):
        self.prev = None
        self.data = None
        self.next = None
```

```python
def add_node(new_node):
    global head, tail
    
    if head == None:
        head = tail = new_node
        return
    tail.next = new_node
    new_node.prev = tail
    tail = new_node

def insert_node(new_node):
    global head, tail
    
    if head == None:
        head = tail = new_node
        return
    
    node = head
    while node:
        if node.data > new_node.data:
            if node == head:  # head앞에 삽입
                new_node.next = head
                node.prev = new_node
                head = new_node
            else:
                node.prev.next = new_node  #  이전노드의 next를  new_node 로 연결
                new_node.prev = node.prev  #   새노드의 이전 주소를 이전 노드로 연결
                node.prev = new_node   # 노드 앞에 추가 노드 연결
                new_node.next = node   # 새 노드 next에 노드 주소 연결
            return
        node = node.next  # 다음 node로 이동
    # tail  에 자료 추가
    tail.next = new_node
    new_node.prev = tail
    tail = new_node

def delete_node(name):
    global head, tail
    
    # name이 있는 node 검색
    node = head
    while node:
        if node.data == name:
            if node == head:
                head.next.prev = None
                head = node.next
            elif node == tail:
                node.prev.next = None
                tail = node.prev
            else:
                node.prev.next = node.next
                node.next.prev = node.prev
            del node
            return True
        node = node.next
    return False
 
def print_node():
    global head
    node = head 
    while node:
        print(node.data)
        node = node.next
    
def reverse_print():
    global tail
    node = tail
    while node:
        print(node.data)
        node = node.prev
```
