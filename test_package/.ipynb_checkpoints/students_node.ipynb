```python
# 학생의 이름, 성적을 입력하여 성적 순으로 싱글링크드 리스트로 자료를 생성
# 1. 학생 정보 입력
# 2. 학생 전체 목록 출력
# 3. 학생 정보 검색   ->  이름을 입력 받아 해당 학생의 이름, 성적을 출력
# 4. 학생 정보 수정   -> 이름을 입력 받아 학생의 정보 출력 후 
#                        점수를 입력 받아 새로운 점수로 수정
# 5. 학생 정보 삭제  -> 이름을 입력받아 검색 한 후 정보 삭제
# 9. 프로그램 종료
```

```python
class Node:
    def __init__(self):
        self.name = None
        self.score = None
        self.link = None
```

```python
def prt_menu():
    menu = '''\
    1. 학생 정보 입력
    2. 학생 전체 목록 출력
    3. 학생 정보 검색  
    4. 학생 정보 수정
    5. 학생 정보 삭제
    9. 프로그램 종료
    
    '''
    print(menu)
    
def insert_node(new_node):   # 마지막에 데이터 추가
    global head
    if head == None:   # 처음 node인 경우, head 에 추가할 node  주소 저장
        head = new_node
        return
    node = head    # 처음  node  지정
    while node.link:     # 추가할 위치 검색 후 node 추가
        if node.score > new_node.score:
            if node == head:    # 제일 앞에 추가할 경우
                head = new_node
                new_node.link = node
            else:  # 중간 node  삽입
                prev.link = new_node   # 이전 node에 새로운 node 연결
                new_node.link = node   # 새로운 node에 현재 node 연결
            return
        prev = node        #  이전 node 저장
        node = node.link   # 다음 node로 이동    
    # 추가할 위치 없음 -> 마지막에 node  추가
    if head == node and node.score > new_node.score:
        head = new_node
        new_node.link = node
    else:
        node.link = new_node  # 마지막 node에 추가

def search_node(name):
    global head
    node = head
    while node:
        if name == node.name:   # 검색할 이름이 node와 같으면 
            return node         # node  반환
        node = node.link   # 다음 자료로 이동
    return 

def update_node(name):
    node = search_node(name)
    if node:
        print(f'{node.name} : {node.score}')        
        score = int(input("성적 입력 "))
        node.score = score
    else:
        print('수정할 자료 없음')
        
def delete_node(name):
    global head
    node = head
    while node:
        if name == node.name:
            if node == head:     # 삭제할 node가 처음 node인 경우
                head = node.link
            else:
                if node.link:   # 삭제할 node가 중간 node인 경우
                    prev.link = node.link
                else:           #  삭제할 node가 마지막 node 인 경우
                    prev.link = None
            del node   #  node 메모리 삭제
            return
        
        prev = node
        node = node.link
        
def print_node():
    node = head
    while node:
        print(f'{node.name} : {node.score}')
        node = node.link    # 다음 자료로 이동  
```

```python
# 1. 학생 정보 입력
# 2. 학생 전체 목록 출력
# 3. 학생 정보 검색   ->  이름을 입력 받아 해당 학생의 이름, 성적을 출력
# 4. 학생 정보 수정   -> 이름을 입력 받아 학생의 정보 출력 후 
#                        점수를 입력 받아 새로운 점수로 수정
# 5. 학생 정보 삭제  -> 이름을 입력받아 검색 한 후 정보 삭제
# 9. 프로그램 종료
head = None
```

```python
while True:
    prt_menu()
    select = int(input("  메뉴 선택 : "))
    if select == 1:
        name, score = input("이름 성적 입력 : ").split()
        score = int(score)
        node = Node()
        node.name = name
        node.score = score
        insert_node(node)
    elif select == 2:
        print_node()
    elif select == 3:
        name = input('검색할 이름 입력 ')
        node = search_node(name)
        if node:
            print(f'{node.name} : {node.score}')
        else:
            print('  자료 없음 ')
    elif select == 4:
        name = input('검색할 이름 입력 ')
        update_node(name)
    elif select == 5:
        name = input('삭제할 이름 입력 ')
        delete_node(name)     
    elif select == 9:
        break                 
```
