```python
# for 반복문과 범위 함수 사용하기
for i in range(5):  # 0, 1, 2, 3, 4
    print(str(i) + " = 반복")
print()

for i in range(5,10):   # 5, 6, 7, 8, 9
    print(str(i) + " = 반복")
print()

for i in range(0,10,3):     #   0, 3, 6, 9
    print(str(i) + " = 반복")
```

```python
array = [273, 32, 103, 57, 52]
for element in array:
    print(element)
print()

for i in range(len(array)):   # index로 리스트의 값을 찾음
    print(f'{i} 번째의 값 : {array[i]}')
print()

for idx, value in enumerate(array):    # 위치 인덱스의 값을 반환
    print(f'{idx} 번째의 값 : {value}')
```

```python
# for 반복문 : 반대로 반복하기
for i in range(4, 0-1, -1):
    print(f'{i} 번째 반복')
print()

for i in reversed(range(5)):
    print(f'{i} 번째 반복')
```

```python
# 중첩 반복문으로 피라미드 만들기
output = ''
for i in range(1,10):
    for j in range(0,i):
        output += '*'
    output += '\n'
print(output)
print()

for i in range(1,10):
    output = ''
    for j in range(0,i):
        output += '*'
    print(output)
print()


output = ''
for i in range(1,10):
    output += ('*'*i)
    output += '\n'
print(output)
print()
```

```python
# 반복문 : while 표현식 :  -> 표현식이 True인 동안 실행
# 반복문 탈출 : break,   반복문 계속 : continue
while True:
    for _ in range(10):
        print('.', end='')
        break
    continue     # while 문장으로 분기
    break      # 실행 안됨
```

```python
list_a = [1,2,3,2,2,5]
value =2

# list_a 에 value가 존재하면 반복문 실행
while value in list_a:   # True인 경우 실행
    list_a.remove(value)
    print("실행 결과 : ", list_a)
print(list_a)
```

```python
# 1부터 100까지의 수 중에서 짝수의 합만 구하세요
# 4의 배수는 제외
# 99이면 반복문 탈출
tot = 0
for i in range(1,100+1):
    if i == 99:
        break
    if i % 4 == 0:
        continue
    if i % 2 == 0:
        tot += i
print(tot)
```

```python
# while 문자형으로 변경
i = 1
tot = 0
# while i <= 100
while True:
    if i == 99:
        break
    if i % 4 ==0:
        i += 1
        continue
    if i % 2 ==0:
        tot += i
    i += 1
print(tot)
```

```python
# 학생의 이름과 성적을 입력받아 딕셔너리에 저장 한 후
# 학생의 이름에 'q'가 입력되면 입력 종료
students = {}
while True:
    student = input("이름과 성적 입력: (예: 홍길동 90 형식, 종료는 q 입력) > ")
    if student == 'q':
        break
    student = student.split()
    if len(student) != 2 or not student[1].isdigit():
        print("다시입력")
        continue
    students[student[0]] = int(student[1])
# 학생의 총 인원 수와 총점, 평균을 구하세요
tot = 0
for key in students:
    tot += students[key]
print('총 인원 수 : {},   총점 : {}, 평균 : {}'.format(len(students), tot, tot/len(students)))
```

```python
# while 반복문 : 시간을 기반으로 반복하기
import time

cnt = 0 # 반복 횟수 초기화

# 5초 동안 반복하기
target_time = time.time() + 5
while time.time() < target_time:
    cnt += 1
print(f'5초 동안 실행한 반복 횟수 : {cnt}')
```

```python
numbers = [102,56,78,12,30]

# numbers 에서 가장 작은 값을 출력
min_num = numbers[0]
for num in numbers[1:]:
    if num< min_num:
        min_num = num
print(" 주어진 리스트에서 가장 작은 값은 : ", min_num)

print(min(numbers))

print(min(34,13,12))
```

```python
# 리스트의 합을 구함
tot = 0
for num in numbers:
    tot += num
print(tot)
print(sum(numbers))
```

```python
# 문) names =[] scores =[]
#1. 학생수 입력
# 입력한 학생 수 만큼 이름과 성적을 입력 받아
# 이름을 키로 성적을 값으로 names, scores 에 저장
names= []
scores =[]
while True:
    input_data = input("이름 성적 입력 : 종료는 q >")
    if input_data == 'q':
        break
    name, score = input_data.split()
    names.append(name)
    scores.append(int(score))
print(names)
print(scores)

# 3. 검색할 이름을 입력받아 입력된 성적 출력, 없으면 '자료 없음' 출력
s_name = input("검색할 이름 입력 > ")
if s_name in names:
    for idx, name in enumerate(names):
        if name == s_name:
            break
    print(f"{s_name} : {scores[idx]}")
else:
    print(' 자료 없음 ')
```
